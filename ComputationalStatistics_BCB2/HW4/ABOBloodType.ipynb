# ABO Blood Type 
#### STEP 0. Define variables and state assumptions

$\begin{array}{rr} \hline
Variable &Definition &Value \\ \hline
X_i &The\ observed\ phenotype\ of\ the\ i^{th}\ individual & \\ \hline
Z_i &The\ hidden\ genotype\ of\ the\ i^{th}\ individual & \\ \hline
\Omega &Phenotype\ Sample\ Space &\{A,\ B,\ O,\ AB\} \\ \hline
K &Genotype\ Possibilities &\{AA,AO,AB,BO,BB,OO\} \\ \hline
N &Number\ of\ individuals\ in\ our\ experiment &100 \\ \hline
N_A,N_B,N_{AB},N_O &Number\ of\ individuals\ with\ given\ phenotype &\{64,\ 10,\ 14,\ 12\} \\ \hline
\theta &Array\ of\ unknown\ parameters  &\{P_A,\ P_O,\ P_B\}
\end{array}$

<p style=""><u>Assumptions</u></p>

1. $X_1,...X_N$ are iid random variables as are hidden data $Z_1,...,Z_N$.
2. The 3 alleles involved with blood phenotype determination are in Hardy-Weinberg equalibream (and all the assumptions HWE makes)
3. Phenotypic observations are correct and there are only 3 alleles that contribute to the phenotype, alleles A, B, and O and each individual has 2 alleles

#### STEP 1. Determine the likelihood

The complete data (Y) likelihood can be written as

$L_C(\theta | Y) = {\displaystyle \prod_{i=1}^{n} L_{Ci}(\theta | X_i,Z_i)}$ 

where $L_Ci(\theta\ |\ X_i,Z_i) = Pr(X_i,Z_i)$ is the complete data likelihood for the ith observed individual. Thus,

$L_C(\theta\ |\ Y) = {\displaystyle \prod_{i=1}^{n}\prod_{k \in K}^{} [P(X_i,Z_i = k]^{\mathbb{1}\{Z_i = k\}}}$ 

The log of $L_C$ is then

$l_C(\theta\ |\ Y) = \displaystyle \sum_{i=1}^{n}\sum_{k \in K}^{}\mathbb{1}\{Z_i = k\}\ ln\ Pr(X_i,\ Z_i\ =\ k)$

#### Step 2. E-Step

$E[l_C(\theta\ |\ y)\ |\ X;\theta^{(m)}] = \displaystyle \sum_{i=1}^{n}\sum_{k \in K}^{}\ E[\mathbb{1}\{Z_i = k\}\ |\ X;\theta^{(m)}]\ ln\ Pr(X_i,\ Z_i\ =\ k)$

The expected value of an indicator for a random variable is the probability of that random variable.

$\ =\ \displaystyle \sum_{i=1}^{n}\sum_{k \in K}^{}\ Pr(Z_i=k\ |\ X_i;\theta^{(m)})\ ln\ Pr(X_i,\ Z_i\ =\ k)$ 

$Pr(Z_i=k\ |\ X_i;\theta^{(m)})$ is computed with the current estimates of $\theta^{(m)}$. Applied to our model for each of the 6 k's $\in$ K

$Pr(Z_i=\ AA|\ X;\theta^{(m)}) = \mathbb{1}\{X_i = A\} \cdot$
$\frac{P_A^{2(m)}}{P_A^{2(m)}+2P_A^{(m)}P_O^{(m)}}$

$Pr(Z_i=\ AO|\ X;\theta^{(m)}) = \mathbb{1}\{X_i = A\} \cdot$
$\frac{2P_A^{(m)}P_O^{(m)}}{P_A^{2(m)}+2P_A^{(m)}P_O^{(m)}}$

$Pr(Z_i=\ BB|\ X;\theta^{(m)}) = \mathbb{1}\{X_i = B\} \cdot$
$\frac{P_B^{2(m)}}{P_B^{2(m)}+2P_B^{(m)}P_O^{(m)}}$

$Pr(Z_i=\ BO|\ X;\theta^{(m)}) = \mathbb{1}\{X_i = B\} \cdot$
$\frac{2P_B^{(m)}P_O^{(m)}}{P_B^{2(m)}+2P_B^{(m)}P_O^{(m)}}$

$Pr(Z_i=\ AB|\ X;\theta^{(m)}) = \mathbb{1}\{X_i = AB\}$

$Pr(Z_i=\ OO|\ X;\theta^{(m)}) = \mathbb{1}\{X_i = OO\}$

The joint probabilities can be expressed for each k as:

$ln\ Pr(X_i,\ Z_i\ =\ AA)\ =\ \mathbb{1}\{X_i = A\} \cdot ln(P_A^2)$

$ln\ Pr(X_i,\ Z_i\ =\ AO)\ =\ \mathbb{1}\{X_i = A\} \cdot ln(2P_AP_O)$

$ln\ Pr(X_i,\ Z_i\ =\ BB)\ =\ \mathbb{1}\{X_i = B\} \cdot ln(P_B^2)$

$ln\ Pr(X_i,\ Z_i\ =\ AO)\ =\ \mathbb{1}\{X_i = B\} \cdot ln(2P_BP_O)$

$ln\ Pr(X_i,\ Z_i\ =\ AB)\ =\ \mathbb{1}\{X_i = AB\} \cdot ln(2P_AP_B)$

$ln\ Pr(X_i,\ Z_i\ =\ OO)\ =\ \mathbb{1}\{X_i = O\} \cdot ln(P_O^2)$

Aggregating like terms and summing over all N:

$Q(\theta |\ \theta^{(m)}) = $

$\ N_A \left(\frac{2P_A^{2(m)}}{P_A^{2(m)}+2P_A^{(m)}P_O^{(m)}}ln(P_A)+\frac{2P_A^{(m)}P_O^{(m)}}{P_A^{2(m)}+2P_A^{(m)}P_O^{(m)}}ln(2P_AP_O)\right) $

$+\ N_B\left(\frac{2P_B^{2(m)}}{P_B^{2(m)}+2P_B^{(m)}P_O^{(m)}}ln(P_B)+\frac{2P_B^{(m)}P_O^{(m)}}{P_B^{2(m)}+2P_B^{(m)}P_O^{(m)}}ln(2P_BP_O)\right)$

$+\ N_{AB}\left(ln(2P_AP_B)\right)$

$+\ N_O\left(2 \ ln(P_O)\right)$



#### Step 3. M-Step
To find $\theta^{(m+1)}$, we took the derivate of the Q function and set it equal to 0 with respect to each of the 3 parameters in $\theta$, namely $(\hat{P_A},\hat{P_B},\hat{P_O})$. To make the partial derivatives possible, we added the Lagrange multiplier to the Q function $Q(\theta |\ \theta^{(m)}) + \lambda(1-P_A-P_B-P_O)$ and set the derivate equal to 0 and solved.
<p style="color:red">See q1.jpg (last page of this pdf) for full derivate steps, otherwise, the results of the derivations are below.</p>

$\hat{P_A} = \frac{\partial Q}{\partial P_A} Q(\theta |\ \theta^{(m)}) + \lambda(1-P_A-P_B-P_O)\ =\ 0$

$=\ \frac{1}{\lambda} \left( 2N_A \left( \frac{P_A^{2(m)}+P_A^{(m)}P_O^{(m)}}{P_A^{2(m)}+2P_A^{(m)}P_O^{(m)}}\right)\right)$

$\hat{P_B} = \frac{\partial Q}{\partial P_B} Q(\theta |\ \theta^{(m)}) + \lambda(1-P_A-P_B-P_O)\ =\ 0$

$=\ \frac{1}{\lambda} \left( 2N_B \left( \frac{P_B^{2(m)}+P_B^{(m)}P_O^{(m)}}{P_B^{2(m)}+2P_B^{(m)}P_O^{(m)}}\right)\right)$

$\hat{P_O} = \frac{\partial Q}{\partial P_O} Q(\theta |\ \theta^{(m)}) + \lambda(1-P_A-P_B-P_O)\ =\ 0$

$=\ \frac{1}{\lambda} \left( 2N_A \left( \frac{2P_A^{(m)}P_O^{(m)}}{P_A^{2(m)}+2P_A^{(m)}P_O^{(m)}} \right) + 2N_B \left( \frac{2P_B^{(m)}P_O^{(m)}}{P_B^{2(m)}+2P_B^{(m)}P_O^{(m)}} \right) +2N_O\right)$

To solve for $\lambda$, we used the known constraint that $P_A + P_B + P_O = 1$.

$\left[ 2N_A[\frac{P_A^{{2}^{(m)}} + P_A^{(m)}P_O^{(m)}}{P_A^{{2}^{(m)}} + 2P_A^{(m)}P_O^{(m)}}] + N_{AB} + 2N_B[\frac{P_B^{{2}^{(m)}} + P_B^{(m)}P_O^{(m)}}{P_B^{{2}^{(m)}} + 2P_B^{(m)}P_O^{(m)}}] + N_{AB} + 2N_A[\frac{P_A^{(m)}P_O^{(m)}}{P_A^{{2}^{(m)}}+2P_A^{(m)}P_O^{(m)}}] + 2N_B[\frac{P_B^{(m)}P_O^{(m)}}{P_B^{{2}^{(m)}}+2P_B^{(m)}P_O^{(m)}}] + 2N_O \right] = \lambda$

After combining like terms:

$[2N_B + 2N_{AB} + 2N_A +2N_O] = \lambda$

$2[N_B + N_{AB} + N_A + N_O] = \lambda$

We obtain $\lambda$:

$ \lambda = 2N $

We now have all the variables we need to make estimations for $\hat{P_A},\hat{P_B},\hat{P_O}$. Although general and scientific evidence exists for the frequency of these alleles in the population, we chose a uniform probabilities for our $m = 0$ initial values. We ran the calculations until the sum of the difference between the 3 parameters of $\theta$ satisfied the following inequality:

$(\theta^{(m+1)} - \theta^{(m)}) < 1e^{-8}$


#### Step 4. Data collection and $\hat{p}$ estimation

```python
from numpy import subtract,float128
data = { # Phenotype observations
    'A' :64,
    'B' :10,
    'AB':14,
    'O' :12,
}

theta_m = { # Initial estimations of P_hat_A, P_hat_B, P_hat_O
    'A':float128(1/3.0),
    'B':float128(1/3.0),
    'O':float128(1/3.0)
}
```

```python
def Q(N, P):
    numAlleles =  float(sum(N.values()) *2)
    
    P_M_A = 1/numAlleles * (2*N['A']*((P['A']**2 #A alleles from AA  
                                     + P['A']*P['O']) #A alleles from AO
                                     /(P['A']**2 + 2*P['A']*P['O'])) #Porportion of A and O alleles
                                     + N['AB']) # Number of A alleles from AB phenotype
    P_M_B = 1/numAlleles * (2*N['B']*((P['B']**2 + P['B']*P['O'])/(P['B']**2 + 2*P['B']*P['O'])) + N['AB'])
    
    P_M_O = 1/numAlleles * (2*N['A']*((P['A']*P['O'])/(P['A']**2 + 2*P['A']*P['O']))   + # Alleles from AO
                              2*N['B']*((P['B']*P['O'])/(P['B']**2 + 2*P['B']*P['O'])) + # Alleles from BO
                              2*N['O'])                                                  # Alleles from OO
    return {'A':P_M_A,'O':P_M_O,'B':P_M_B}
                             
```

```python
dif = 1 # Initialize the difference between the 2 theta vectors to be greater than the cuttoff so the loop will run
m = 1 #The mth EM estimation
cutoff = float128(1.0*10**-8) # cuttoff value for when to stop the EM
while dif > cutoff:
    theta_m_plus_1 = Q(data,theta_m)
    dif = sum(abs(subtract(list(theta_m.values()),list(theta_m_plus_1.values()))))
    print("%i. %s" % (m,str(theta_m_plus_1)))
    m+=1
    theta_m = theta_m_plus_1
```
